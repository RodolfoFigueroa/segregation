```python
import os
import pickle
from pathlib import Path

import pandas as pd
```

```python
segregation_path = Path(os.getenv("SEGREGATION_PATH"))
```

```python
df = {}
for path in (segregation_path / "results").glob("*.pkl"):
    with open(path, "rb") as f:
        data = pickle.load(f)
        df[path.stem] = data["H"].item()

df = pd.Series(df)
```

```python
df.plot.bar()
```
